```python
import csv
from common import *
```

```python
positive_images = []
negative_images = []

with open('./data-labels/images.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        image_name = row['IMAGE FILENAME'].strip()
        is_positive = 1 if row['IS POSITIVE'] else 0
        if not image_name:
            break
        image_path = './data-images/{}'.format(image_name)
        if is_positive:
            positive_images.append(image_path)
        else:   
            negative_images.append(image_path)

total_count = len(positive_images) + len(negative_images)
```

## All Positives:

```python
print("Count {} / {} = {:.2f}%".format(len(positive_images), total_count, 100 * len(positive_images) / total_count))
display_images(positive_images)
```

## All Negatives:

```python
print("Count {} / {} = {:.2f}%".format(len(negative_images), total_count, 100 * len(negative_images) / total_count))
display_images(negative_images)
```
